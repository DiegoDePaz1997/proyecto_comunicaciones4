```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = 19 #orden del filtro

    
    
def sinc(orden_filtro):
    orden_filtro = L
    h = np.ones(L)
    n = np.array(range(-L//2 + 1, L//2 + 1))    
    w = np.linspace(-np.pi, np.pi, 5*L)#para verla mejor agregar mas puntos
    F = lambda w: np.sum(h*np.exp(w*n*1j))
    f = np.array([F(i)for i in w])/(2*np.pi)

    return f

def w(unos):
    return np.ones(unos*5)


f = np.convolve(sinc(L), w(L), 'full')

f = abs(f)
plt.figure(figsize = (15,5))
plt.plot(f)
```

```python

def sinc2(M):
    w= 100
    n = np.linspace(0, M-1, M)
    s = n - M//2
    
    s[(M-1)//2] = 2
    lista =np.sin(w*s)/(np.pi*s)
    lista[(M-1)//2] = 1.1*lista[(M-1)//2 + 1]            
    return lista










orden = 91
h = np.convolve(sinc2(orden), np.ones(orden), 'full')
h = abs(h)
plt.plot(h)
plt.plot(sinc2(orden))


```

```python
def potencia(num):
    numeros = [i for i in range(2,int(num**.5)+1) if num%i == 0]
    for g in numeros:
        for i in range(1,int(num**.5)+2):
            if(num == g**i): return [g, i]
    return None
            
        
        
x = [potencia(i) for i in range(1, 100) if potencia(i)!= None]        
print(x)
list(map(lambda a: a[0]**a[1], x))
```
