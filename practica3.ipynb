```python
import numpy as np
import matplotlib.pyplot as plt

L = 11

h = np.ones(L)
n = np.array(range(-L//2 + 1, L//2 + 1))
w = np.linspace(-np.pi, np.pi, 10*L)#para verla mejor agregar mas puntos


#para sacar TFTD -pi<w<pi
F = lambda w: np.sum(h*np.exp(-w*n*1j))


f = [F(i).real for i in w]






plt.figure(figsize= (15,5))
plt.subplot(1,2, 1)
plt.plot(f)

plt.subplot(1,2, 2)
plt.stem(n, h);

```

```python
l = 800
lon = np.linspace(-l//2, l//2 + 1, l)
t = np.linspace(0, 3, l)
fun = np.sin(100*2*np.pi*t) + .8*np.sin(300*2*np.pi*t) + 8*2*np.random.random(l) - 8*1

fftt = np.fft.fft(fun, norm='forward')
fftt = np.abs(fftt)

plt.figure(figsize = (15, 8))
plt.subplot(2,1,1)
plt.plot(fun);
plt.subplot(2,1,2)
L = len(fftt)
plt.plot(lon, fftt)
plt.grid()
```

```python
f_muestreo = 1000
lista_corr = []

def senal(f_senal):    
    t = np.arange(0, 50, 1)
    X = np.cos(2*np.pi*t*f_senal/f_muestreo)
    return X


def conv(X,Y):
    corr = np.convolve(X, Y[::-1], 'full')
    lista_corr.append(max(corr))
    plt.stem(corr);


#autocorrelacion
X = senal(100)
conv(X, X)



#correlacion cruzada 100, 200, 300, 400
plt.figure(figsize=(15,8))

for i in range(100, 500, 100):
    plt.subplot(2,2,i//100)
    plt.title(f'{i}Hz')
    Y = senal(i)
    conv(X, Y)
    
    
lista_corr= [round(100*i/lista_corr[0],3) for i in lista_corr]    
for i, j in zip(lista_corr, [100,100,200,300,400]):
    print(f'{j}Hz: {i} %')
  

    






```

```python
#DFT
x = np.array([2,-1,0,4,0,-1])# entrada
#k es un numero entre 0 y 5 para este x
def DFT_K(k, x):
    num = np.array(list(range(len(x))))
    lista = x*np.exp(-1j*2*np.pi*k*num/len(x))
    print(lista)    
    lista = sum(lista)

    return lista
```
