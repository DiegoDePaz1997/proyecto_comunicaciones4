# Convoluci√≥n

```python
import matplotlib.pyplot as plt 
import numpy as np 

plt.figure(figsize=(24,5))
muestras = 4
unos = 2


#funcion muestreada
v = np.linspace(-10, muestras - 1 + 5, muestras + 10 + 5)
n = np.zeros(len(v))
n[2:unos + 2]= 1
 

#funcion exp
rr = np.linspace(0, muestras - 1, muestras)
y = 2*np.exp(-0.1*rr)
y = np.array([1,1,1,1])
n[10: muestras + 10] = y
#funcion escalon
#y = np.ones(muestras)

#vector h de convolucion
h = np.array([1,1])


plt.subplot(1, 2, 1)
plt.stem(v, n)


#la convolucion
convolucion = np.convolve(h, y, mode='full')
plt.subplot(1, 2, 2)
plt.stem(convolucion);

plt.subplots_adjust(left=0.1,
                    bottom= .1,
                    right=1,
                    top=1,#longitud vertical cuadro entero
                    wspace=0.1,#espacio entre plots vertical
                    hspace=0.3)#espacio entre plots horizontal


plt.show()
```

#### Remuestreo

resample es combinacion de las dos anteriores
la funcion hace el diesmado e interpolado a la vez

```python
plt.figure(figsize = (12, 12))

muestras = 50
n = np.linspace(0, muestras - 1, muestras)
y = np.sin(2*np.pi*0.05*n)

f = signal.resample_poly(y, 5, 15)# 5 = arriba  15 = abajo


plt.subplot(2, 1, 1)
plt.stem(y)
plt.subplot(2, 1, 2)
plt.stem(f);
```
