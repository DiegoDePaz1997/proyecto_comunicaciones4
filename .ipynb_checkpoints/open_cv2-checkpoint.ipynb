```python
import cv2
imagen = cv2.imread('yo.png', 1)#cv2.IMREAD_GRAYSCALE
h, w = imagen.shape[:2]
print(f'altura= {h}  ancho= {w}')
```

```python
'''cada punto de la imagen es una tupla de los colores, en orden blue green red'''
(B, G, R) = imagen[100, 100]
print(f'R = {R} G= {G} B= {B}')

#se puede acceder a cada valor individual de color si se especifica
B = imagen[100, 100, 0]
print(f'B = {B}')
```

```python
pequena = cv2.resize(imagen, (0, 0), fx=.5, fy=.5)
cv2.imshow('imagen', pequena)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
grande = cv2.resize(imagen, (0, 0), fx=2, fy=2)
cv2.imshow('imagen', grande)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.subplot(1, 2, 1)
plt.imshow(imagen);
plt.subplot(1, 2, 2)
plt.imshow(pequena);
```

```python
edges = cv2.Canny(imagen, 100, 150)
cv2.imshow('edges', edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
'''deteccion de bordes'''
```

```python
gaussiano = cv2.GaussianBlur(imagen, (17, 17), 0)
cv2.imshow('gausiano', gaussiano)
cv2.waitKey(0)
cv2.destroyAllWindows()
'''desenfoque gausiano'''
```

```python
def decorador_factorial(f):
    diccionario = {}
    
    def annky(arg):
        if(arg not in diccionario.keys()):
            print(f'no estaba {arg}')
            diccionario[arg] = f(arg)
        return diccionario[arg]
    return annky

@decorador_factorial
def factorial(num):
    if(num == 1):
        return 1
    else:
        return num*factorial(num - 1)
```

```python
factorial(5)
```
